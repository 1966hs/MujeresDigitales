```python
def f(x): #funcion exponencial, x es un vector de numpy
    return np.exp(-x ** 2) #hacemos el vector y el exponente

x = np.linspace(-5,5, num=1000) #creamos numeros de -5 hasta 5 con mil idvisiones 

plt.plot(x,f(x), label="f(x) Function")
plt.ylabel('f(x)')#nombre del eje y 
plt.xlabel('x')
plt.legend()#esto es la ubicacion de recuadro pqueno 
```

```python
plt.plot(x,f(x), 'k-.', label="f(x) Function") # k es para que la linea sea negra y punteada . label recordemos que es el titulo
plt.ylabel('f(x)')#nombre eje y
plt.xlabel('x')#nombre del eje x
plt.legend()#recuadro pqueno
```

```python
x = np.linspace(-10,10, num=10) #desde -10 hasta 10 
plt.plot(x,f(x), linestyle='--', marker='o', label="f(x) Function") #diferentes marcadores para los puntos que estamos graficando 
plt.ylabel('f(x)')
plt.xlabel('x')
plt.legend()
```

```python
#ESTUDIAR ESTE EJERCICIO 

x = np.arange(0,10)
y = np.arange(0,10)

median = 3

plt.plot(x,y, label="Linear")
plt.ylabel('Y')
plt.xlabel('X')
plt.legend(loc='upper center')
plt.fill_between(x,y,median, where=(y > median), interpolate=True, alpha=0.25)# X,Y WHERE DONDE LA Y SEA MAYOR QUE LA MEDIANA DE AZUL
plt.fill_between(x,y,median, where=(y <= median), interpolate=True, color='red', alpha=0.25)#DONDE Y SEA MENOR QUE LA MEDIANA SE PONE DE COLOR RED
```

```python
x = np.arange(0,10)
y =  np.arange(0,10)
y2 = [0,1,1,2,3,6,7,7,5,6]

plt.plot(x,y, label="Linear")
plt.plot(x,y2, label="Linear 2")
plt.ylabel('Y')
plt.xlabel('X')
plt.fill_between(x,y,y2, where=(y > y2), interpolate=True, alpha=0.25, label='Above y2')
plt.fill_between(x,y,y2, where=(y <= y2), interpolate=True, color='red', alpha=0.25, label='Below y2')
plt.legend(loc='upper left')
```
